# Inspecting and Replacing missing values

```python
# Inspect missing values in the dataset
print(df.isnull().values.sum())

# Replace the ' 's with NaN
df = df.replace(" ",np.NaN)


# Impute the missing values with mean imputation
df = df.fillna(df.mean())

# Count the number of NaNs in the dataset to verify
print(df.isnull().values.sum())
```

## Our EDA cycle starts from here.

```python
#Visualize Correlations in a heatmap

plt.figure(figsize=(8,8))
sns.heatmap(df.corr(), annot=True)
```

## Experience Levels of Data Science Jobs

```python
plt.figure(figsize=(10,6))
plt.title('Experience Levels of Data Science Jobs', fontsize=18)
ax = sns.countplot (x='experience_level', data=df)
for i in ax.containers:
    ax.bar_label(i)
plt.xlabel('Experience Level', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.show()
```

**This bar graph shows the the experience level of Data Science Employee**
where
1. EN means Entry-Level/Junior
2. SE means Senior-level/Expert
3. EX means Executive-level/Director
4. MI means Mid-level/Intermediate 

## Job Titles in Data Science

```python
plt.figure(figsize=(15,15))
plt.title('Various Job Titles in Data Science', fontsize=20, fontweight='bold')
sns.countplot(y='job_title', data=df, order=df['job_title'].value_counts().index)
plt.xlabel('Count', fontsize=15, fontweight='bold')
plt.ylabel('Job Titles', fontsize=20, fontweight='bold')
plt.show()
```

**Maximum number of employees are working as Data Scientist.**

**whereas the count of Financial Data Analyst, Product Data Analysis, Applied Machine Learning Scientist, Finance Data Analyst, Product Data Analyst, 3D Computer Vision Researcher, Staff Data Scientist, Applied Data Scientist, Data Architect, Marketing Data Analyst, Bid Data Analyst, Principle Data Analyst, Data Specialist is less.**

## Salary in USD

```python
plt.figure(figsize=(5,5))
sns.boxplot(y='salary_in_usd', data=df)
plt.show()
```

## Distribution plot of Data Science Salaries in USD

```python
plt.figure(figsize=(11,6))
sns.distplot(df['salary_in_usd'])
plt.title('Distribution plot of Data Science Salaries in USD', fontsize=18)
plt.show()
```

***This plot shows an increase in the number of Data Science jobs and salary for each year.***

* Right skewness: mean > median 

## Distribution of salary by work year

```python
plt.figure(figsize=(10,5))
sns.kdeplot(data=df, x='salary_in_usd', hue='work_year', fill=True)
plt.title("Distribution of salary by work year", fontsize=16)
plt.xlabel("Salary (in USD)", fontsize=14)
plt.ylabel("Density", fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```

***The density plot shows an increase in the number of Data Science jobs and salary for each year.***

## Highest salary by Job Title

```python
title_sal_pivot = pd.pivot_table(data=df,index=['job_title'],values=['salary_in_usd'], aggfunc='max').sort_values(by=['salary_in_usd'],ascending=False)
display(title_sal_pivot)
```

## Highest Salaries by Job Titles in Data Science

```python
plt.figure(figsize=(15,15))
plt.title('Highest Salaries by Job Titles in Data Science', fontsize=20, fontweight='bold')
sns.lineplot(y='salary_in_usd', x='job_title', data=df)
plt.xlabel('Job Titles', fontsize=15, fontweight='bold')
plt.ylabel('Amount', fontsize=20, fontweight='bold')
plt.xticks(rotation=90)
plt.show()
```

## Employee Residence and Company Location of Data Science

```python
fig,axes=plt.subplots(1,2,figsize=(20,15))
sns.countplot(ax=axes[0],y='employee_residence', data=df, order=df['employee_residence'].value_counts().index)
sns.countplot(ax=axes[1],y='company_location', data=df,  order=df['company_location'].value_counts().index)

axes[0].set_title('Employee Residence', fontsize=25, fontweight='bold')
axes[0].set(xlabel=None)
axes[0].set(ylabel=None)
axes[1].set_title('Company Location', fontsize=25, fontweight='bold')
axes[1].set(xlabel=None)
axes[1].set(ylabel=None)

fig.supxlabel('Count', fontsize=20, fontweight='bold')
fig.supylabel('Location', fontsize=20, fontweight='bold')

plt.show()
```

**For Employee Residence and Company location, the US is the highest in terms of frequency, followed by India ,Denmark and Great Britain. Whereas the least amount of Employee Residence and Company location are in Mexico, Singapore,Phillipines.**
